## Reward Curves for fedML




```python
import numpy as np
import pandas as pd

# Hypothetical scenario assumptions

# Assumed base emission for the network, which is a pool of available tokens to be distributed monthly
base_emission = 100000  # M
# Assumed network utilization rate, which reflects the proportion of nodes that are actively participating
utilization_rate = 0.75  # U, 75% of nodes are active
# Assumed individual node stake relative to the total network stake (hypothetical)
node_stake_ratio = 0.01  # S, 1% of the total stake in the network
# Assumed individual node reputation score, NOTE THAT WE NORMALIZE HERE, WILL MAKE MORE SENSE LATER
node_reputation_score = 0.85  # R, normalized between 0 and 1


# formula for a node's earnings
def calculate_node_earnings(M, U, S, R):
    return (M * (1 - U) * S) + (M * U * R)

# Calculate node earnings with the provided hypothetical scenario
node_earnings_original = calculate_node_earnings(base_emission, utilization_rate, node_stake_ratio, node_reputation_score)

```

```python
# Define a range of utilization rates, node stake ratios and reputation scores to analyze

# this helps us understand how nodes respond to different levels of activity, stake, and reputation
utilization_rates = np.array([0.25, 0.50, 0.75, 1.00])  # Different levels of network activity
node_stake_ratios = np.array([0.01, 0.05, 0.10, 0.20])  # From low to very high stake
node_reputation_scores = np.array([0.1, 0.5, 0.85, 1.0])  # From very low to perfect reputation

# calculate node earnings for each combination of utilization rate, stake ratio, and reputation score
earnings_data = []
for u in utilization_rates:
    for s in node_stake_ratios:
        for r in node_reputation_scores:
            original_earnings = calculate_node_earnings(base_emission, u, s, r)
            earnings_data.append({
                'Utilization Rate': u,
                'Stake Ratio': s,
                'Reputation Score': r,
                'Original Earnings': original_earnings,
            })

# covert to df for better visualization
earnings_df = pd.DataFrame(earnings_data)

earnings_df.sort_values(by=['Utilization Rate', 'Stake Ratio', 'Reputation Score'], inplace=True)
earnings_df.reset_index(drop=True, inplace=True)

earnings_df.head(50)
```

## Simulating Reward Curve with different Reputation, Utilization, and Stakes


```python
# calculating reputation score

import numpy as np
import pandas as pd

# hypothetical scenario assumptions
base_emission = 100000  # M
utilization_rate = 0.75  # U, 75% of nodes are active
node_stake_ratio = 0.01  # S, 1% of the total stake in the network


# Reputation Score
# R = (days of month deployed/total days of month) + ∑(contributed revenue / num of nodes in deployment)

def calculate_reputation_score(days_deployed, total_days_in_month, contributed_revenues, num_nodes):
  """
  Calc reputation score for a single node

  params:
  """
  active_ratio = days_deployed / total_days_in_month
  revenue_score = np.sum(np.array(contributed_revenues) / np.array(num_nodes))
  return active_ratio * revenue_score



# calculation for node earnings
def calculate_node_earnings(M, U, S, R):
    return (M * (1 - U) * S) + (M * U * R)
```

```python
# observing reputation score behavior

import matplotlib.pyplot as plt
total_days_in_month = 30


scenarios = {
    'Scenario 1': {
        'days_deployed': 30,
        'contributed_revenues': [1000, 1500, 2000],
        'num_nodes': [10, 15, 20]
    },
    'Scenario 2': {
        'days_deployed': 15,
        'contributed_revenues': [800, 1200],
        'num_nodes': [8, 12]
    },
    'Scenario 3': {
        'days_deployed': 25,
        'contributed_revenues': [500, 700, 900, 100],
        'num_nodes': [5, 7, 9, 1]
    }
}

results = {}
for name, params in scenarios.items():
    r_score = calculate_reputation_score(
        params['days_deployed'],
        total_days_in_month,
        params['contributed_revenues'],
        params['num_nodes']
    )
    results[name] = r_score

labels, scores = zip(*results.items())
plt.bar(labels, scores)
plt.xlabel('Scenario')
plt.ylabel('Reputation Score')
plt.title('Reputation Scores for Different Scenarios')
plt.show()

results
```

## Simulations

```python
# simulations

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def calculate_reputation_score(days_deployed, total_days_in_month, contributed_revenues, num_nodes):
    active_ratio = days_deployed / total_days_in_month
    revenue_score = np.sum(np.array(contributed_revenues) / np.array(num_nodes))
    return active_ratio * revenue_score

def calculate_node_earnings(M, U, S, R):
    return (M * (1 - U) * S) + (M * U * R)

def calculate_monthly_emissions(base_annual_emission, demand_factor, offset):
    ncm = max(min(demand_factor - offset, 1), -1)
    base_monthly_emission = base_annual_emission / 12
    monthly_emission = base_monthly_emission * (1 + ncm)
    return monthly_emission

# simulation parameters
num_simulations = 1000
num_months = 12
total_emission = 10_000_000_000 * 0.15
base_annual_emission = total_emission / 5
example_offset = 0.10

# generate random scenarios for each simulation
utilization_rates = np.random.uniform(0.2, 0.9, num_simulations)
node_stake_ratios = np.random.uniform(0.01, 0.2, num_simulations)
total_days_in_month = 30

# Run sims
node_earnings_data = []
for i in range(num_simulations):
    U = utilization_rates[i]
    S = node_stake_ratios[i]

    for month in range(num_months):
        demand_factor = np.random.uniform(0.5, 1.5)
        base_emission = calculate_monthly_emissions(base_annual_emission, demand_factor, example_offset)

        days_deployed = np.random.randint(1, total_days_in_month + 1)
        num_deployments = np.random.randint(1, 6)
        contributed_revenues = np.random.uniform(100, 2000, num_deployments)
        num_nodes = np.random.randint(1, 21, num_deployments)

        R = calculate_reputation_score(days_deployed, total_days_in_month, contributed_revenues, num_nodes)
        node_earnings = calculate_node_earnings(base_emission, U, S, R)

        node_earnings_data.append({
            'Simulation': i,
            'Month': month,
            'Utilization Rate': U,
            'Stake Ratio': S,
            'Reputation Score': R,
            'Node Earnings': node_earnings,
            'Demand Factor': demand_factor,
            'Base Emission': base_emission
        })


node_earnings_df = pd.DataFrame(node_earnings_data)

# viz sim results
summary_stats = node_earnings_df.groupby('Simulation').agg({
    'Node Earnings': ['mean', 'min', 'max'],
    'Reputation Score': 'mean',
    'Utilization Rate': 'mean',
    'Stake Ratio': 'mean'
})
print("Summary Statistics:")
print(summary_stats)

# viz distribution of node earnings
plt.figure(figsize=(10, 6))
plt.hist(node_earnings_df['Node Earnings'], bins=50, edgecolor='black')
plt.xlabel('Node Earnings')
plt.ylabel('Frequency')
plt.title('Distribution of Node Earnings')
plt.show()

# sensitivity analysis
# varying utilization rate
utilization_rates_sensitivity = np.linspace(0.2, 0.9, 5)
sensitivity_results = []
for u in utilization_rates_sensitivity:
    sensitivity_df = node_earnings_df[np.abs(node_earnings_df['Utilization Rate'] - u) < 1e-2]
    print(f"Utilization Rate: {u}")
    print(f"Number of rows in sensitivity_df: {len(sensitivity_df)}")
    if len(sensitivity_df) > 0:
        mean_earnings = sensitivity_df['Node Earnings'].mean()
        sensitivity_results.append({
            'Utilization Rate': u,
            'Mean Node Earnings': mean_earnings
        })
    else:
        print("No data found for this utilization rate.")
sensitivity_df = pd.DataFrame(sensitivity_results)

print("Sensitivity Analysis Results:")
print(sensitivity_df)

plt.figure(figsize=(10, 6))
plt.plot(sensitivity_df['Utilization Rate'], sensitivity_df['Mean Node Earnings'], marker='o')
plt.xlabel('Utilization Rate')
plt.ylabel('Mean Node Earnings')
plt.title('Sensitivity Analysis - Utilization Rate')
plt.show()

# viz long-term trends
monthly_earnings = node_earnings_df.groupby('Month')['Node Earnings'].mean()

plt.figure(figsize=(10, 6))
plt.plot(monthly_earnings.index, monthly_earnings.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Mean Node Earnings')
plt.title('Long-term Trend of Node Earnings')
plt.show()
```

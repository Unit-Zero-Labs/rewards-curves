## Monthly Emission Rates

```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_monthly_emissions(base_annual_emission, demand_factor, offset):
    """
    Calculate the monthly emissions based on annual emissions and network computation adjustments.

    :param base_annual_emission: The total emissions to be distributed annually
    :param demand_factor: Factor by which demand adjusts the base emission. A measure of total computational resources demanded from the network
    :param offset: Adjustment to the demand factor to calculate the NCM if necessary
    :return: The monthly emission rate for the current month
    """
    # Calculate NCM directly using demand_factor and offset, ensuring it is between -1 and 1
    ncm = max(min(demand_factor - offset, 1), -1)

    # Calculate the base monthly emission without multiplier
    base_monthly_emission = base_annual_emission / 12

    # Adjust base monthly emission based on NCM
    monthly_emission = base_monthly_emission * (1 + ncm)
    return monthly_emission

# Total STEMS to be emitted over 5 years is 15% of 10,000,000,000
total_emission = 10_000_000_000 * 0.15
base_annual_emission = total_emission / 5

# Example parameters
example_offset = 0.10  # Example offset

# Calculate monthly emissions with different demand factors from 0.5 to 1.5
demand_factors = np.linspace(0.5, 1.5, 10)
monthly_emissions_results = []
for demand_factor in demand_factors:
    emission_rate = calculate_monthly_emissions(base_annual_emission, demand_factor, example_offset)
    monthly_emissions_results.append(emission_rate)

plt.plot(demand_factors, monthly_emissions_results, marker='o')
plt.title('Monthly Emissions vs. Demand Factor')
plt.xlabel('Demand Factor')
plt.ylabel('Monthly Emissions (STEMS)')
plt.grid(True)
```
